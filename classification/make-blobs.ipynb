# Make Blobs Classification

## Load Dataset

```python
from sklearn.datasets import make_blobs

X, y = make_blobs(random_state=42)
```

```python
print("X shape:", X.shape)
print("y shape:", y.shape)
```

```python
X[:5]
```

```python
import mglearn
from matplotlib import pyplot as plt

mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.legend(["Class 0", "Class 1", "Class 2"])
plt.show()
```

## Split Dataset into Train and Test

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

## Train and Evaluate Model

### Support Vector Machine

```python
from sklearn.svm import LinearSVC

linear_svm_model = LinearSVC()
linear_svm_model.fit(X_train, y_train)

print("Training set score: {:.3f}".format(linear_svm_model.score(X_train, y_train)))
print("Test set score: {:.3f}".format(linear_svm_model.score(X_test, y_test)))
```

```python
print("Coefficient shape: ", linear_svm_model.coef_.shape)
print("Intercept shape: ", linear_svm_model.intercept_.shape)
```

```python
# Visualization
import numpy as np

mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
line = np.linspace(-15, 15)
for coef, intercept, color in zip(linear_svm_model.coef_, linear_svm_model.intercept_,
                                  mglearn.cm3.colors):
    plt.plot(line, -(line * coef[0] + intercept) / coef[1], c=color)
plt.ylim(-10, 15)
plt.xlim(-10, 8)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.legend(['Class 0', 'Class 1', 'Class 2', 'Line class 0', 'Line class 1', 'Line class 2'],
           loc=(1.01, 0.3))
plt.show()
```

```python
# Visualization
mglearn.plots.plot_2d_classification(linear_svm_model, X, fill=True, alpha=.7)
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
line = np.linspace(-15, 15)
for coef, intercept, color in zip(linear_svm_model.coef_, linear_svm_model.intercept_,
                                  mglearn.cm3.colors):
    plt.plot(line, -(line * coef[0] + intercept) / coef[1], c=color)
plt.legend(['Class 0', 'Class 1', 'Class 2', 'Line class 0', 'Line class 1', 'Line class 2'],
           loc=(1.01, 0.3))
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.show()
```
